```python
def triangle(n):
    for i in range(1, 7):
        if i == 1:
            print((7 - i) * " " + "*")
        elif i <= 5:
            print((7 - i) * " " + "*" + (2 * i - 3) * " " + "*")
        if i == 6 :
            print((6-i) *" ", i*"* ")
triangle(7)
```
