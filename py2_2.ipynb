```python
def triangle(n):
    for i in range(1, n):
        if i == 1:
            print((n - i) * " " + "*")
        elif i <= n-2:
            print((n - i) * " " + "*" + (2 * i - 3) * " " + "*")
        if i == n-1 :
            print((n-1-i) *" ", i*"* ")
triangle(7)
```

```python
def triangle(n):
    for i in range(n,0,-1):
        print((n-i) *" ", (i+(i-1))*"*")
    for i in range(2,n+1):
        print((n-i) *" ", (i+(i-1))*"*")
triangle(7)
```

```python
def triangle(n):
    for i in range(1,n):
        print((n-1-i) *" ", i*"* ")
    for i in range(n-2,0,-1):
        print((n-1-i) *" ", i*"* ")
triangle(7)
```
